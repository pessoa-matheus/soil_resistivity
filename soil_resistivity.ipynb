```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
spacing = np.linspace(0.1,1)
def function(x):
    a = np.linspace(0.1,1)
    p = [0] * x
    f = [0] * x
    p[0] = 0.5
    for i in range(x):
        p[i] = p[i-1] + 0.1
    for i in range(x):
        f1 = 2*a/(np.sqrt(a**2 + 4*p[i]**2))
        f2 = a/(np.sqrt(a**2 + p[i]**2))
        f[i] =  4*np.pi*a/(1+ (f1-f2))
    
    return f 
a,b,c,d,e,f,g,h,i,j = function(10)

plt.plot(spacing, a, label = 'p=0.1')
plt.plot(spacing, b, label = 'p=0.2')
plt.plot(spacing,c, label = 'p=0.3')
plt.plot(spacing,d, label = 'p=0.4')
plt.plot(spacing,e, label = 'p=0.5')
plt.plot(spacing, f, label = 'p=0.6')
plt.plot(spacing, g, label = 'p=0.7')
plt.plot(spacing,h, label = 'p=0.8')
plt.plot(spacing,i, label = 'p=0.9')
plt.plot(spacing,j, label = 'p=1.0')
plt.xlabel('Spacing [m]')
plt.ylabel('Geometrical factor')
plt.legend()
plt.savefig('geom.png')
plt.show()
```
